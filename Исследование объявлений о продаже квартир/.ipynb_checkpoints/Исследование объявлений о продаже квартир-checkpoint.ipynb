# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

# Изучение данных из файла

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data_raw = pd.read_csv('/datasets/real_estate_data.csv', sep='\t')
pd.set_option('max_columns', None)
pd.set_option('display.float_format', '{:.1f}'.format)

data_raw.head()
```

Удалим столбцы, которые не понядобятся в данном исследовании - они делают вывод датафрейма громоздким и замусоренным

```python
data = data_raw.drop(columns=['total_images',
                              'is_apartment',
                              'airports_nearest',
                              'parks_around3000',
                              'parks_nearest',
                              'ponds_around3000',
                              'ponds_nearest',
                              ])
data.info()
print('\nКол-во дублей:', data.duplicated().sum(), '\n')
data.describe()
```

#### Выявленные проблемы:
- Пропуски во множестве столбцов.


- Неверные типы данных в нескольких столбцах


- Выбросы вроде 100 метрового полотка

```python
report = data.isna().sum().to_frame()
report = report.rename(columns = {0: 'missing_values'})
report['% of total'] = ((report['missing_values'] / data.shape[0])*100).round(2)
report.sort_values(by = 'missing_values', ascending = False)
```

# Предобработка данных

#### Пойдем по порядку: пропуски данных.

- В столбцах с количеством этажей и названиями населеных пунктов невозможно заполнить пропуски, потому такие значения будут удалены - их достаточно мало для такого.


- Высоту полотков заполним медианным значением.


- Столбец с балконами можно заполнить нулями, потому что отсутствие данных по ним скорее всего означает отсутствие балконов.


- Столбцы с площадями жилой области и кухни считаю возможным заполнить медианами с группировкой по общей площади.


- Столбец срока жизни объявления заполним медианой.


- Расстояние до центра заполнить невозможно без нужных данных, так что оставляем без изменений

```python
# Удаляем строки с пустыми значениями

data.dropna(subset=['locality_name', 'floors_total'], inplace=True)
data = data.reset_index(drop=True)
```

```python
# Заполняем пустые строки медианами и нулями

data['ceiling_height'].fillna(data['ceiling_height'].median(), inplace=True)
data['days_exposition'].fillna(data['days_exposition'].median(), inplace=True)

data['balcony'].fillna(0, inplace=True)
```

```python
# Разбиваем общую площать на группы с помощью квартилей

area_data = data[['total_area', 'living_area', 'kitchen_area']].copy()
area_data['area_group'] = pd.qcut(x=area_data['total_area'], q=4, labels=['Малая', 'Средняя', 'Выше среднего', 'Большая'])
area_data_grouped = area_data.pivot_table(index='area_group', values=['living_area', 'kitchen_area'], aggfunc='median')
area_data_grouped
```

```python
# Заполняем пропуски в датафрейме медианными значениями групп
for group_name in area_data_grouped.index:
    data.loc[data['living_area'].isna(), 'living_area'] = area_data_grouped['living_area'][group_name]
    
    data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = area_data_grouped['kitchen_area'][group_name]
```

#### Производим замену типов данных:

- Цену продажи, кол-во балконов и этажей, расстояние до центра и время жизни объявления можно оставить как есть, пото


- Дату объявления меняем со строки на datetime

```python
# Спикок столбцов, которые перейдут в целочисленные
# Последние два перевожу из 64 в 32 бита просто для красоты
type_list = ['last_price', 'floors_total', 'balcony', 'days_exposition', 'rooms', 'floor']
data[type_list] = data[type_list].astype(int)

data['cityCenters_nearest'] = data['cityCenters_nearest'].astype("Int32")

data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

# Проверяем пропуски и замену типов
data.info()
```

#### Вывод
Датасет готов к работе

```python
data.loc[data['floor'] == 1, 'floor_category'] = 'Первый'
data.loc[data['floor'] == data['floors_total'], 'floor_category'] = 'Последний'
data['floor_category'] = data['floor_category'].fillna('Другой')
```

```python
data['living_total_area_ratio'] = data['living_area'] / data['total_area']
data['kitchen_total_area_ratio'] = data['kitchen_area'] / data['total_area']
```

## Исследовательский анализ данных

В данных много выбросов, которые исказят визуализацию.
Прежде, чем строить графики и делать какие-то выводы, нужно избавиться от выбросов


```python
# Напишем функцию, которая определяет границы нормальных значений данных
def normal_data_size(column):
    stat = data[column].describe()
    iqr = stat[6] - stat[4]
    
    minimal = round(stat[4] - iqr * 1.5, 1)
    maximal = round(stat[6] + iqr * 1.5, 1)
    
    if minimal < stat[3]:
        minimal = stat[3]
        
    if maximal > stat[7]:
        maximal = stat[7]
    
    return [minimal, maximal]
```

#### Изучаем данные площади, цены, комнат и потолков
С помощью функции просматриваем границы нормальных значений и выводим гистограммы, исходя из этого диапазона

```python
normal_data_size('total_area')
```

```python
normal_data_size('last_price')
```

```python
data.plot(y='last_price', kind='hist', bins=50, grid=True, title='Цена продажи')
plt.legend().remove()
plt.ylabel('Количество квартир')
plt.xlabel('Цена продажи')
plt.show()
```

#### Уберем выбивающиеся значения из датафрейма

Создадим новые столбцы, с помощью которых позже отсеим выбросы

```python
data['big_area'] = data['total_area'] > normal_data_size('total_area')[1]
data['expencive'] = data['last_price'] > normal_data_size('last_price')[1]
data['much_rooms'] = data['rooms'] > normal_data_size('rooms')[1] 
data['strange_ceiling'] = ((data['ceiling_height'] < normal_data_size('ceiling_height')[0]) 
                         | (data['ceiling_height'] > normal_data_size('ceiling_height')[1]))
data.head()
```

Создадим новый датафрейм, в котором отсеим строки с выбросами.

Сразу же можно удалить эти столбцы, потому что далее они не нужны, свою функцию они выполнили.

```python
clean_data = data[(data['big_area'] == False) &
                  (data['expencive'] == False) &
                  (data['much_rooms'] == False) &
                  (data['strange_ceiling'] == False)
                 ]
clean_data.reset_index(drop=True)

clean_data = clean_data.drop(columns=[
                            'big_area',
                            'expencive',
                            'much_rooms',
                            'strange_ceiling'
                            ])
```

## Общий вывод

Для определения рыночной стоимости недвижимости стоит обращать внимание на следующие параметры:
- Общая площадь. Выше площадь - выше стоимость.
- Близость к центру города. Ближе к центру - выше стоимость.
- Высота потолков. Выше потолки - выше стоимость.
- Этаж. Первый или последний этажи - ниже стоимость.
- Количество комнат. Больше комнат - выше стоимость.
