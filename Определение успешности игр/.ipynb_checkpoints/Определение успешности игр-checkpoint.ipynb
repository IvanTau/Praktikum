# Определение успешности игр

Интернет-магазин "Стримчик", продающий видеоигры, просит провести анализ предоставленных данных и найти закономерности, влияющие на продажи видеоигр, с целью определения рекламной стратегии на предстоящий год.

# Изучение данных

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st
```

```python
data = pd.read_csv('/datasets/games.csv')
display(data.head(3))
data.describe()
```

```python
print('Количество дубликатов:', data.duplicated().sum())
print('Количество пропущенных значений:')
display(data.isna().sum())

print('Типы данных:')
data.info()
```

**Выявленные проблемы:**
- Заголовки в uppercase
- Множество пропущенных значений
- Неверный тип данных в трех столбцах
- Огромное число игр с нулевыми продажами, в их числе ряд популярных игр, что вызывает **сомнения в достоверности датасета**.

**Возможные причины пропусков**

Легко предположить, что пропуски в оценках критиков и юзеров вызваны тем, игры были выпущены задолго до появления подобных оценок и платформ для них. Однако, огромное число пропусков можно наблюдать и в свежих играх. Потому можем обвинить в пропусках ошибки сбора и выгрузки данных

# Предобработка данных
Пропущенные значения я решил не заполнять ничем, поскольку считаю, что это исказит результаты анализа в дальнейшем. 
Удалять строки с пропусками так же не буду, за исключением двух строк с пропусками названия и жанра игр

```python
# Меняем заголовки на lowercase
data.columns = data.columns.str.lower()

# Удалим две строки с отсутствующими значениями имени и жанра
data.dropna(subset=['name'], inplace=True)
data = data.reset_index(drop=True)

# Приведем столбцы к нужным типам данных
data.year_of_release = data.year_of_release.astype('Int32')
data.critic_score = data.critic_score.astype('Int32')

data.user_score = data.user_score.replace('tbd', None)
data.user_score = data.user_score.astype(float)

data.info()
```

```python
# Добавим столбец с суммой всех продаж
data['total_sales'] = data.na_sales + data.eu_sales + data.jp_sales
data.head()
```

# EDA

Посмотрим на количество выпущенных игр за каждый год

```python
sns.set_theme(style='darkgrid', palette='tab10', font_scale=1.2)
```

```python
plt.figure(figsize=(8, 4))
ax = sns.histplot(data=data, x='year_of_release', binwidth=1)
ax.set(ylabel='Кол-во игр за год', xlabel='Год выпуска')
plt.show()
```

На графике виден рост количества выпускаемых игр за год вплоть до 2010 года, затем падение.


```python
# Найдем платформы с наибольшими суммарными продажами и построим распределение продаж
platform_top = data.pivot_table(index=['platform'], values='total_sales', aggfunc='sum')\
    .sort_values(by='total_sales', ascending=False).head(6)

plt.figure(figsize=(10, 5))
for platform in platform_top.index:
    query = data.query('platform == @platform').pivot_table(
        index='year_of_release', values='total_sales', aggfunc='sum')
    plt.plot(query, 'o-',  label=platform)


plt.xlabel('Год')
plt.ylabel('Продажи за год')
plt.legend()
plt.show()

platform_top
```

На графике видим, что в среднем платформы живут примерно 10 лет, пик продаж по каждой длится около 5 лет.

За актуальный период возьмем 2011-2015. 2016-й год не станем включать, так как за него данные неполные и из-за этого на графиках наблюается падение продаж

```python
data_fresh = data.query('2011 <= year_of_release <= 2015')
```

```python
# Найдем лидирующие платформы за актуальный период и построим распределение продаж
platform_top_fresh = data_fresh.pivot_table(index=['platform'], values='total_sales', aggfunc='sum')\
    .sort_values(by='total_sales', ascending=False).head(6)

plt.figure(figsize=(10, 5))
for platform in platform_top_fresh.index:
    query = data_fresh.query('platform == @platform').pivot_table(
        index='year_of_release', values='total_sales', aggfunc='sum')
    plt.plot(query, 'o-',  label=platform)


plt.xlabel('Год')
plt.ylabel('Продажи за год')
plt.legend()
plt.show()

platform_top_fresh
```

На графике наблюдается рост продаж у платформ Playstation 4 и Xbox One, у других платформ продажи падают, что вполне логично - две лидирующие платформы явлюятся более свежими по сравнению с другими.

```python
# Построим бокплот по продажам на популярных платформах за актумальный период
data_top = data_fresh.query('platform in @platform_top_fresh.index')

plt.figure(figsize=(16, 5))
sns.boxplot(data=data_top, x='total_sales', y='platform')
plt.xlim(0, 2.5)
plt.xlabel('Продажи')
plt.ylabel('Платформа')
plt.show()
```

На графике видно огромное число выбросов на всех платформах, распределения сдвинуты ближе к нулю, максимумы не превышают 2.5 млн. копий. ХOne лидирует по медианным продажам, PS4 дышит ему в спину, однако на прошлом графике видно лидерство PS4 по общему числу продаж.

###### Посмотрим на влияние оценок критиков и юзеров на продажи на платформе PS3 - у нее больше всего проданных копий за актуальный период

```python
data_ps3 = data_top.query(
    'platform == "PS3" and (critic_score > 0 or user_score > 0)')

plt.figure(figsize=(10, 5))
sns.scatterplot(data=data_ps3, x='critic_score', y='total_sales')
plt.ylabel('Продажи')
plt.xlabel('Рейтинг критиков')
plt.show()
print('Корреляция продаж и рейтинга критиков', data_ps3['critic_score'].astype('float64').corr(data_ps3['total_sales']), '\n')

plt.figure(figsize=(10, 5))
sns.scatterplot(data=data_ps3, x='user_score', y='total_sales')
plt.ylabel('Продажи')
plt.xlabel('Рейтинг юзеров')
plt.show()
print('Корреляция продаж и рейтинга юзеров', data_ps3['user_score'].astype('float64').corr(data_ps3['total_sales']))
```

Наблюдается небольшая корреляция продаж и рейтинга критиков, в то же время корреляция с рейтингом юзеров практически отсутствует.

###### Посмотрим на другие платформы:

```python
plt.figure(figsize=(10, 5))
for platform in platform_top_fresh.index:
    query = data_top.query(
        'platform == @platform and (critic_score > 0 or user_score > 0)')[['total_sales', 'critic_score']]
    sns.scatterplot(data=query, x='critic_score',
                    y='total_sales', legend=False)

plt.legend(labels=platform_top_fresh.index)
plt.ylabel('Продажи')
plt.xlabel('Рейтинг критиков')
plt.show()
print('Корреляция продаж и рейтинга критиков', data_top['critic_score'].astype('float64').corr(data_top['total_sales']), '\n')

plt.figure(figsize=(10, 5))
for platform in platform_top_fresh.index:
    query = data_top.query(
        'platform == @platform and (critic_score > 0 or user_score > 0)')[['total_sales', 'user_score']]
    sns.scatterplot(data=query, x='user_score', y='total_sales', legend=False)

plt.legend(labels=platform_top_fresh.index)
plt.ylabel('Продажи')
plt.xlabel('Рейтинг юзеров')
plt.show()
print('Корреляция продаж и рейтинга юзеров', data_top['user_score'].astype('float64').corr(data_top['total_sales']))
```

В целом картина аналогичная: Небольшая корреляция с критиками и практически нулевая корреляция с юзерами.

В целом у японцев прослеживается пристрастие к карманным консолям, в то время, как американцы и европейцы предпочитают стационарные варианты.

Скорее всего это вызвано тем, что японцы не всегда могут себе позволить дорогие стационарные консоли с телевизорами как в плане денег, так и в плане свободного места в маленьких квартирах.
Так же они, скорее всего, более активно пользуются общественным транспортом, чтобы добираться до работы, нежели американцы и европейцы, а ехать долго на автобусе довольно скучно.

# Проверка гипотез

###### Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

H0 - Рейтинги не отличаются.

H1 - Рейтинги отличаются.

```python
xone = data_fresh.query('platform == "XOne"')['user_score'].dropna()
pc = data_fresh.query('platform == "PC"')['user_score'].dropna()

alpha = 0.05

results = st.ttest_ind(xone, pc)

print('Средний рейтинг XOne:', xone.mean())
print('Средний рейтинг PC:', pc.mean(), '\n')
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

***
###### Гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

H0 - Рейтинги не отличаются.

H1 - Рейтинги отличаются.

```python
action = data_fresh.query('genre == "Action"')['user_score'].dropna()
sports = data_fresh.query('genre == "Sports"')['user_score'].dropna()

alpha = 0.05

results = st.ttest_ind(action, sports)

print('Средний рейтинг Action:', action.mean())
print('Средний рейтинг Sports:', sports.mean(), '\n')
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Как мы видим, тесты говорят в пользу приведенных гипотез.
